```python
import googlemaps
import pandas as pd
import gmaps as gs
```

```python
gmaps = googlemaps.Client(key='YOUR KEY HERE')
gs.configure(api_key='YOUR KEY HERE') 
```

# TechPoint

```python
techpoint = gmaps.geocode('1210 Waterway Blvd Suite 5000 Indianapolis, IN 46202')
techpoint_loc = techpoint[0]['geometry']['location']
techpoint_loc
```

```python
techpoint_marker = (techpoint_loc['lat'],techpoint_loc['lng'])
x = gs.figure(center=(techpoint_loc['lat'],techpoint_loc['lng']), zoom_level=12)
markers = gs.marker_layer([techpoint_marker])
x.add_layer(markers)
x
```

# Find A Place To Stay

```python
hotel = gmaps.places_nearby(location=techpoint_loc, 
                             radius=2000,
                             keyword = 'hotel',
                             #rank_by='distance',
                             type='lodging')

hotels = pd.DataFrame(hotel['results'])
```

```python
hotels.head()
```

```python
#Visualize Hotels
hotels['geo'] = hotels.apply(lambda x: (x['geometry']['location']['lat'],x['geometry']['location']['lng']), axis=1)
hotel_layer = gs.symbol_layer(list(hotels['geo']), fill_color='yellow', 
                              stroke_color='yellow', scale=10,hover_text = list(hotels['name']) )
x.add_layer(hotel_layer)
x
```

```python
hotels = hotels[hotels['rating']>4.4]
hotels
```

```python
hotel_layer = gs.symbol_layer(list(hotels['geo']), fill_color='black', 
                              stroke_color='black', scale=10,hover_text = list(hotels['name']) )
x.add_layer(hotel_layer)
x
```

# Things To Do

```python
uni = gmaps.places_nearby(location=techpoint_loc, 
                             radius=5000,
                             keyword = 'university',
                             #rank_by='distance',
                             type='university')
museum = gmaps.places_nearby(location=techpoint_loc, 
                             radius=5000,
                             type='museum')
restaurant = gmaps.places_nearby(location=techpoint_loc, 
                             radius=5000,
                             type='restaurant')
```

```python
restaurant2 = gmaps.places_nearby(page_token = restaurant['next_page_token'])
```

```python
unis = pd.DataFrame(uni['results'])
museums = pd.DataFrame(museum['results'])
restaurants = pd.DataFrame(restaurant['results']).append(pd.DataFrame(restaurant2['results']))

```

```python
restaurants.head()
```

## Identifying location Data for All Destinations

```python
unis['geo'] = unis.apply(lambda x: (x['geometry']['location']['lat'],x['geometry']['location']['lng']), axis=1)
museums['geo'] = museums.apply(lambda x: (x['geometry']['location']['lat'],x['geometry']['location']['lng']), axis=1)
restaurants['geo'] = restaurants.apply(lambda x: (x['geometry']['location']['lat'],x['geometry']['location']['lng']), axis=1)
```

## Layer Markers in Google Maps

```python
x = gs.figure(center=(techpoint_loc['lat'],techpoint_loc['lng']), zoom_level=12)
markers = gs.marker_layer([techpoint_marker])
x.add_layer(markers)

unis_layer = gs.symbol_layer(list(unis['geo']), fill_color='red', stroke_color='red', scale=10,hover_text = list(unis['name']) )
museums_layer = gs.symbol_layer(list(museums['geo']), fill_color='green', stroke_color='green', scale=10,hover_text = list(museums['name']) )
restaurants_layer = gs.symbol_layer(list(restaurants['geo']), fill_color='blue', stroke_color='blue', scale=10,hover_text = list(restaurants['name']) )

x.add_layer(unis_layer)
x.add_layer(museums_layer)
x.add_layer(restaurants_layer)
x
```
